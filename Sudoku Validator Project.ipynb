#Sudoku Validator using NumPy
This project validates a Sudoku board by checking :
- Row sum
- Column sum
- 3x3 grid sum
If all conditions satisfy Sudoku rules,it prints "Valid Sudoku"

```python
import numpy as np
def is_valid_sudoku(sudoku):
  #check row sum
  for i in range(sudoku.shape[0]):
    if sudoku[i].sum() !=45:
      return"Inavlid Sudoku"

  #check column sum
  for j in range(sudoku.shape[1]):
    if sudoku[:,j].sum() !=45:
      return "Invalid Sudoku"

  #check 3x3 grid sum
  for i in range(0,9,3):
    for j in range(0,9,3):
      if np.sum(sudoku[i:i+3,j:j+3]) !=45:
        return"Invalid Sudoku"

  return"Valid Sudoku"


```

```python


sudoku = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])

print(is_valid_sudoku(sudoku))
```
